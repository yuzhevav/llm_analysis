```python
from mastodon import Mastodon
import pandas as pd
```

```python
token = '*'

mastodon = Mastodon(
    access_token=token,
    api_base_url='https://mastodon.itiabd.online'
)
```

```python
limit = 40
all_toots = []
max_id = None

while True:
    public_toots = mastodon.timeline_public(limit=limit, max_id=max_id)

    if not public_toots:
        break

    all_toots.extend(public_toots)

    # Обновляем max_id для следующей итерации
    max_id = public_toots[-1]['id']  # Берем ID последнего полученного поста
```

```python
len(all_toots)
```

```python
all_toots_df = pd.DataFrame(all_toots)
all_toots_df.shape
```

```python
all_toots_df.to_csv('all_toots_df.csv', index=False)
```

```python
text_toots = [toot['content'] for toot in all_toots]
```

```python
# Чистим первые 160 постов, которые не относятся к llm
text_toots = text_toots[:len(text_toots) - 160]
```

```python
# Посты с упоминанием других пользователей
mention_toots = [toot for toot in text_toots if 'class="u-url mention">' in toot]
```
